```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# crime_data = pd.read_csv("data-files/05.crime_in_seoul.csv", encoding="utf-8")
crime_data = pd.read_csv("data-files/05.crime_in_seoul.csv", encoding="utf-8", index_col=0)
crime_data
```

```python
pd.pivot_table(crime_data, index="구별", aggfunc=np.sum)
```

```python
original_crime_data = crime_data.copy()
```

```python
crime_data = pd.pivot_table(crime_data, index="구별", aggfunc=np.sum)
crime_data.head()
```

```python
crime_data["강간검거율"] = crime_data["강간 검거"] / crime_data["강간 발생"] * 100
```

```python
crime_data.columns
```

```python
columns1 = np.array(crime_data.columns)
columns2 = ["강간검거율", "강도검거율", "살인검거율", "절도검거율", "폭력검거율"]
```

```python
for idx, column in enumerate(columns2):
    crime_data[column] = crime_data[columns1[idx * 2]] / crime_data[columns1[idx * 2 + 1]] * 100
```

```python
crime_data.drop(["강간 검거", "강도 검거", "살인 검거", "절도 검거", "폭력 검거"],
                axis=1, inplace=True)
crime_data
```

```python
crime_data.rename(columns={"강간 발생" : "강간",
                           "강도 발생" : "강도",
                           "살인 발생" : "살인",
                           "절도 발생" : "절도",
                           "폭력 발생" : "폭력"}, inplace=True)
crime_data.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
columns3 = ["강간", "강도", "살인", "절도", "폭력"]
data = crime_data[columns3].values #dataframe 데이터 -> ndarray 데이터

scaler = MinMaxScaler()
# x_scaled = scaler.fit(data).transform(data)
x_scaled = scaler.fit_transform(data)
scaled_crime_data = pd.DataFrame(x_scaled, columns=columns3, index=crime_data.index)

columns4 = ["강간검거율", "강도검거율", "살인검거율", "절도검거율", "폭력검거율"]
scaled_crime_data[columns4] = crime_data[columns4]

scaled_crime_data
```

```python
cctv_data = pd.read_csv("data-files/05.cctv.csv", encoding="utf-8", index_col="구별")
cctv_data
```

```python
scaled_crime_data[["인구수", "CCTV"]] = cctv_data[["인구수", "소계"]]
scaled_crime_data
```

```python
# scaled_crime_data["범죄"] = np.sum(scaled_crime_data[columns3], axis=1)
# np.sum(scaled_crime_data[columns3], axis=1)
scaled_crime_data["범죄"] = scaled_crime_data[columns3].sum(axis=1)
scaled_crime_data.head()
                  
```

```python
scaled_crime_data["검거율"] = scaled_crime_data[columns4].sum(axis=1)
scaled_crime_data.head()
```

```python
sorted_scaled_crime_data = scaled_crime_data.sort_values(by="검거율", ascending=False)
plt.figure(figsize=(10, 10))
sns.heatmap(sorted_scaled_crime_data[columns4], annot=True, fmt="f" cmap="RdPu")
plt.show()
```
